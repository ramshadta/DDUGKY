# Handling Null Values

```python
data=pd.read_csv('Windspeed.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

## Filling a common value to all missing fields

```python
# let's try filling 0 to all data
data.fillna(0)
```

## Adding missing data to individual columns

#### In the same way we can add missing data to variouse columns differently, just need to pass values as dictionaries

```python
data.fillna({
    'Temperature':0,
    'Windspeed':5,
    'Status':'sunny'
})
```

## Forward fill (row)
#### The forward fill is a method to forward data from the row above the missing value.

```python
data.fillna(method='ffill')
```

## Backward fill (row)
#### similiar to forwaed fill , it will add missing data from row bottom of missing field

```python
data.fillna(method='bfill')
```

## Limiting the forward/backward fill
#### we can limit the number of rows or columns getting filled

```python
data.fillna(method='ffill',limit=1)
```

## Filling with pandas objects
#### There are pandas objects like sum(),max(),mean(),ets. we can fill missing values with these too

```python
data.fillna(data.mean())
```

## Filling specific range of columns

```python
data.fillna(data.median()['Temperature':'Windspeed'])
```

## Interpolate missing values

```python
data.interpolate()
```

```python
data.Date.interpolate(method='time')
```
