```python
#library installation
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset
df=pd.read_csv('titanic.csv')
df.head()
```

```python
#data type of each columns
df.info()
```

```python
# Number of passengers in each class
df.Pclass.value_counts()
```

## Single group single column

```python
# Number of passengers who survived from each class
class_survived=df.groupby(['Pclass'])['Survived']
class_survived.sum()
```

### Multiple columns

```python
# total amount paid by each class who survived
class_survived_fare=df.groupby(['Pclass'])['Survived','Fare']
class_survived_fare.sum()
```

### Multiple Groups

```python
# count of male and female who survived from each class
survived_gender_count=df.groupby(['Pclass','Sex'])['Survived']
survived_gender_count.sum()
```

```python
# in other way
survived_gender_count1=df.groupby(['Pclass','Sex'],as_index=False)['Survived']
survived_gender_count1.sum()
```

### Multiple Functions

```python
# mean and sum of Pclass and suvived
class_survived_mean_sum=df.groupby(['Pclass'])['Survived'].agg(['mean','sum'])
class_survived_mean_sum
```

### pd.cut method

```python
#creating bins for age
df['Age_bins']=pd.cut(df.Age,bins=3,labels=('Young','Middle','Old'))
df.head()
```

```python
#number of survived from each boarding station
boarded_survived=df.groupby(['Boarded'])['Survived']
boarded_survived.sum()
```
