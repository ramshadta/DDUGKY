# PLOTING WITH PANDAS

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

```

```python
df=pd.read_csv('Car_sales.csv')
df.sample(10)
```

### Droping NaN values

```python
df.dropna(inplace=True)
df.info()
```

## Plots on numerical value columns
### Histogram

```python
df.Horsepower.plot.hist(figsize=(6,3),cmap='brg');
```

### Box plot

```python
df.Sales_in_thousands.plot.box(figsize=(4,3),cmap='gist_earth');
```

```python
df.Horsepower.plot.box(figsize=(4,3),cmap='Blues_r');
```

### Subplots

```python
df.plot.box(figsize=(22,4),subplots=True);
```

```python
df.plot.hist(figsize=(6,18),subplots=True);
```
